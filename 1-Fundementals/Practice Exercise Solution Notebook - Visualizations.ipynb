# Tips Dataset


`Data Features:`

**total_bill**: a numeric vector, the bill amount (dollars)


**tip**: a numeric vector, the tip amount (dollars)


**sex**: a factor with levels Female Male, gender of the payer of the bill


**smoker**: a factor with levels No Yes, whether the party included smokers


**day**: a factor with levels Friday Saturday Sunday Thursday, day of the week


**time**: a factor with levels Day Night, rough time of day


**size**: a numeric vector, number of people in a table/Party

### Q3] Understand the distribution of 'tip' and 'total_bill' using different plots

Hint: Use distplot and boxplot

```python
sns.set({'figure.figsize': (8,6)})
sns.distplot(tips['total_bill'],color='hotpink')
plt.show()
sns.set({'figure.figsize': (8.9,6)})
sns.boxplot(tips['total_bill'],orient='v')
plt.show()
```

```python
sns.distplot(tips['tip'],color='lime')
plt.show()
sns.boxplot(tips['tip'],orient='v')
plt.show()
```

`Both 'tip' and 'total_bill' are skewed towards right but 50 % people paying less than 20 dollars for overall bill and 50 % people paying less than 3 dollars as tip`

### Q4] Relationship between 'total_bill' and 'tip'

Implement jointplot, hex jointplot, pairplot and lmplot

```python
sns.jointplot(tips['total_bill'], tips['tip'],color='green')
plt.show()
```

```python
sns.jointplot(tips['total_bill'], tips['tip'], kind = "hex", color='red')
plt.show()
```

```python
sns.pairplot(tips[['total_bill','tip']])
plt.show()
```

```python
sns.lmplot(x = "total_bill",y = "tip", data = tips)
plt.show()
```

`A linear realtionship can be observed as the bill amount increases so does the tip`

### Q5] Make plots to check distribution of tip with different gender. 

Use striplot and boxplot

```python
sns.stripplot(tips['sex'], tips['tip'])
plt.show()
```

```python
sns.boxplot(tips['sex'], tips['tip'])
plt.show()
```

`Median value for tip is almost same for both, but male have a lot of outliers giving higher tip`

### Q6] Draw relationship between 'total_bill', 'tip' and 'sex'

use lmplot

```python
sns.lmplot(x='total_bill',y='tip',data=tips,fit_reg=False,hue='sex')
plt.show()
```

`Outliers with higher billing amount and tip are males`

### Q7] Which day brings in higher bill amount?

use stripplot, swarmplot and boxplot

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
plt.show()

#shows distribution of data points as well
```

```python
sns.boxplot(y='total_bill',x='day',data=tips)
plt.show()
```

`'Sat' and 'Sun' seems to be good days as total bill and number of customers are high on those days`

### Q8] Find out count of total males and females visiting each day.

### Q9-B] At what time of which day does the restaurant get higher amount of tip (consider for male and female separately)
